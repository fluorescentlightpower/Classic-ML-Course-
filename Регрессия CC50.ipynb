```python
import pandas as pd
import numpy as np
import optuna
from sklearn.preprocessing import PowerTransformer, StandardScaler

from sklearn.pipeline import Pipeline

from sklearn.model_selection import KFold, cross_val_score
from sklearn.linear_model import ElasticNet
from sklearn.ensemble import RandomForestRegressor, ExtraTreesRegressor
from lightgbm import LGBMRegressor
from sklearn.metrics import r2_score, mean_absolute_error, root_mean_squared_error

from google.colab import files
import joblib
```

```python
data = joblib.load('CC50_dataset_regression.joblib')

features = data["features"]
X_train = data["X_train"].loc[:, features]
X_test = data["X_test"].loc[:, features]
y_train = data["y_train"]
y_test = data["y_test"]
pt = data["pt"]

best_metrics = []
best_metrics_orig = []
```

```python
X_train
```

Построим несколько моделей. Линейная - ElasticNet (так как потенцильно высока мультиколлинеарность), случайный лес - Random Forest и Extra Trees, а также градиентный бустинг LightGBM. Оптимизацию гиперпараметров выполняем на кросс-валидации с помощью optuna - более осмысленный перебор, чем GridSearchCV.

```python
cv = KFold(n_splits=5, shuffle=True, random_state=42)

def objective_elasticnet(trial):
    params = {
        "alpha": trial.suggest_float("alpha", 1e-5, 10.0, log=True),
        "l1_ratio": trial.suggest_float("l1_ratio", 0.0, 1.0),
        "max_iter": 100000,
        "random_state": 42
    }

    model = Pipeline([
        ("scaler", StandardScaler()),
        ("elasticnet", ElasticNet(**params))
    ])

    scores = cross_val_score(
        model,
        X_train,
        y_train,
        scoring="neg_root_mean_squared_error",
        cv=cv,
        n_jobs=-1
    )

    return scores.mean()
```

```python
study = optuna.create_study(direction="maximize")
study.optimize(objective_elasticnet, n_trials=50)

print("Лучшая RMSE при кросс-валидации:", -study.best_value)
print("Оптимальные параметры:", study.best_params)
```

```python
best_elasticnet = Pipeline([
    ("scaler", StandardScaler()),
    ("elasticnet", ElasticNet(
        **study.best_params,
        max_iter=100000,
        random_state=42
    ))
])

best_elasticnet.fit(X_train, y_train)

y_pred = best_elasticnet.predict(X_test)

y_test_orig = pt.inverse_transform(
    y_test.reshape(-1, 1)
).ravel()

y_pred_orig = pt.inverse_transform(
    y_pred.reshape(-1, 1)
).ravel()

best_metrics_orig.append({
    "model": "ElasticNet",
    "R2": r2_score(y_test_orig, y_pred_orig),
    "MAE": mean_absolute_error(y_test_orig, y_pred_orig),
    "RMSE": root_mean_squared_error(y_test_orig, y_pred_orig)
})

best_metrics.append({
    "model": "ElasticNet",
    "R2": r2_score(y_test, y_pred),
    "MAE": mean_absolute_error(y_test, y_pred),
    "RMSE": root_mean_squared_error(y_test, y_pred)
})

print("R2:", best_metrics[-1]["R2"])
print("MAE:", best_metrics[-1]["MAE"])
print("RMSE:", best_metrics[-1]["RMSE"])
```

```python
def objective_extratree(trial):
    params = {
        "n_estimators": trial.suggest_int("n_estimators", 100, 1000),
        "max_depth": trial.suggest_int("max_depth", 2, 30),
        "min_samples_split": trial.suggest_int("min_samples_split", 2, 30),
        "min_samples_leaf": trial.suggest_int("min_samples_leaf", 1, 20),
        "max_features": trial.suggest_float("max_features", 0.3, 1.0),
        "bootstrap": trial.suggest_categorical("bootstrap", [True, False]),
        "random_state": 42,
        "n_jobs": -1
    }

    model = ExtraTreesRegressor(**params)

    scores = cross_val_score(
        model,
        X_train,
        y_train,
        scoring="neg_root_mean_squared_error",
        cv=cv,
        n_jobs=-1
    )

    return scores.mean()
```

```python
study = optuna.create_study(direction="maximize")
study.optimize(objective_extratree, n_trials=50)

print("Лучшая RMSE при кросс-валидации:", -study.best_value)
print("Оптимальные параметры:", study.best_params)
```

```python
best_extratree = ExtraTreesRegressor(
    **study.best_params,
    random_state=42,
    n_jobs=-1
)

best_extratree.fit(X_train, y_train)

y_pred = best_extratree.predict(X_test)

y_test_orig = pt.inverse_transform(
    y_test.reshape(-1, 1)
).ravel()

y_pred_orig = pt.inverse_transform(
    y_pred.reshape(-1, 1)
).ravel()

best_metrics_orig.append({
    "model": "Extra Trees",
    "R2": r2_score(y_test_orig, y_pred_orig),
    "MAE": mean_absolute_error(y_test_orig, y_pred_orig),
    "RMSE": root_mean_squared_error(y_test_orig, y_pred_orig)
})

best_metrics.append({
    "model": "Extra Trees",
    "R2": r2_score(y_test, y_pred),
    "MAE": mean_absolute_error(y_test, y_pred),
    "RMSE": root_mean_squared_error(y_test, y_pred)
})

print("R2:", best_metrics[-1]["R2"])
print("MAE:", best_metrics[-1]["MAE"])
print("RMSE:", best_metrics[-1]["RMSE"])
```

```python
def objective_forest(trial):
    params = {
        "n_estimators": trial.suggest_int("n_estimators", 100, 800),
        "max_depth": trial.suggest_int("max_depth", 2, 20),
        "min_samples_split": trial.suggest_int("min_samples_split", 2, 30),
        "min_samples_leaf": trial.suggest_int("min_samples_leaf", 1, 20),
        "max_features": trial.suggest_float("max_features", 0.3, 1.0),
        "bootstrap": trial.suggest_categorical("bootstrap", [True, False]),
        "random_state": 42,
        "n_jobs": -1
    }

    model = RandomForestRegressor(**params)

    scores = cross_val_score(
        model,
        X_train,
        y_train,
        scoring="neg_root_mean_squared_error",
        cv=cv,
        n_jobs=-1
    )

    return scores.mean()
```

```python
study = optuna.create_study(direction="maximize")
study.optimize(objective_forest, n_trials=50)

print("Лучшая RMSE при кросс-валидации:", -study.best_value)
print("Оптимальные параметры:", study.best_params)
```

```python
best_forest = RandomForestRegressor(
    **study.best_params,
    random_state=42,
    n_jobs=-1
)

best_forest.fit(X_train, y_train)

y_pred = best_forest.predict(X_test)

y_test_orig = pt.inverse_transform(
    y_test.reshape(-1, 1)
).ravel()

y_pred_orig = pt.inverse_transform(
    y_pred.reshape(-1, 1)
).ravel()

best_metrics_orig.append({
    "model": "Random Forest",
    "R2": r2_score(y_test_orig, y_pred_orig),
    "MAE": mean_absolute_error(y_test_orig, y_pred_orig),
    "RMSE": root_mean_squared_error(y_test_orig, y_pred_orig)
})

best_metrics.append({
    "model": "Random Forest",
    "R2": r2_score(y_test, y_pred),
    "MAE": mean_absolute_error(y_test, y_pred),
    "RMSE": root_mean_squared_error(y_test, y_pred)
})

print("R2:", best_metrics[-1]["R2"])
print("MAE:", best_metrics[-1]["MAE"])
print("RMSE:", best_metrics[-1]["RMSE"])
```

```python
def objective_lightgbm(trial):
    params = {
        "n_estimators": trial.suggest_int("n_estimators", 100, 800),
        "learning_rate": trial.suggest_float("learning_rate", 0.005, 0.1, log=True),
        "num_leaves": trial.suggest_int("num_leaves", 4, 32),
        "max_depth": trial.suggest_int("max_depth", 2, 6),
        "min_child_samples": trial.suggest_int("min_child_samples", 10, 80),
        "subsample": trial.suggest_float("subsample", 0.5, 1.0),
        "colsample_bytree": trial.suggest_float("colsample_bytree", 0.4, 1.0),
        "bagging_fraction": trial.suggest_float("bagging_fraction", 0.4, 1.0),
        "bagging_freq": trial.suggest_int("bagging_freq", 1, 7),
        "reg_alpha": trial.suggest_float("reg_alpha", 1e-4, 10.0, log=True),
        "reg_lambda": trial.suggest_float("reg_lambda", 1e-4, 10.0, log=True),
        "min_split_gain": trial.suggest_float("min_split_gain", 0.0, 1.0),
        "min_child_weight": trial.suggest_int("min_child_weight", 1, 10),
        "random_state": 42,
        "verbose": -1,
    }

    model = LGBMRegressor(**params)

    scores = cross_val_score(
        model,
        X_train,
        y_train,
        scoring="neg_root_mean_squared_error",
        cv=cv,
        n_jobs=-1
    )

    return scores.mean()
```

```python
study = optuna.create_study(direction="maximize")
study.optimize(objective_lightgbm, n_trials=50)

print("Лучшая RMSE при кросс-валидации:", -study.best_value)
print("Оптимальные параметры:", study.best_params)
```

```python
best_lightgbm = LGBMRegressor(
    **study.best_params,
    random_state=42,
    verbose=-1
)

best_lightgbm.fit(X_train, y_train)

y_pred = best_lightgbm.predict(X_test)

y_test_orig = pt.inverse_transform(
    y_test.reshape(-1, 1)
).ravel()

y_pred_orig = pt.inverse_transform(
    y_pred.reshape(-1, 1)
).ravel()

best_metrics_orig.append({
    "model": "LightGBM",
    "R2": r2_score(y_test_orig, y_pred_orig),
    "MAE": mean_absolute_error(y_test_orig, y_pred_orig),
    "RMSE": root_mean_squared_error(y_test_orig, y_pred_orig)
})

best_metrics.append({
    "model": "LightGBM",
    "R2": r2_score(y_test, y_pred),
    "MAE": mean_absolute_error(y_test, y_pred),
    "RMSE": root_mean_squared_error(y_test, y_pred)
})

print("R2:", best_metrics[-1]["R2"])
print("MAE:", best_metrics[-1]["MAE"])
print("RMSE:", best_metrics[-1]["RMSE"])
```
